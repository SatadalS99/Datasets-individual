# Rock vs Mine prediction (Sonar Data)

```python
#importing the libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

## Data collecting and data processing

```python
#loading the dataset to a pandas dataframe
print('Summary Statistics of actual DataFrame are as follows:')
print('-'*100)
sonar_df = pd.read_csv(r'C:\Users\satad\Documents\Data sets\Rock mine sonar\archive\Sonar_data.csv', header=None)
sonar_df.head(10)
```

```python
#shape of the dataset
print('Summary Statistics of shape of the DataFrame are as follows:')
print('-'*100)
sonar_df.shape
```

```python
#data type information
print('Summary Statistics of datatype information of the DataFrame are as follows:')
print('-'*100)
sonar_df.info()
```

```python
#finding null values
print('Summary Statistics of null values of the DataFrame are as follows:')
print('-'*100)
sonar_df.isnull().sum()
```

```python
#statistical measures of the data
print('Summary Statistics of statistical measures of the DataFrame are as follows:')
print('-'*100)
sonar_df.describe()
```

```python
#checking column index to find number of 
print('Summary Statistics of categorical features of the DataFrame are as follows:')
print('-'*100)
sonar_df[60].value_counts()
```

```python
#mean values of each column
print('Summary Statistics of mean values for each column of the DataFrame are as follows:')
print('-'*100)
sonar_df.groupby(60).mean()
```

```python
#separating the data and lables
print('Summary Statistics of separated data(X) and labels(Y) of the DataFrame are as follows:')
print('-'*100)
X = sonar_df.drop(columns=60, axis = 1)
Y = sonar_df[60]
print(X)
print(Y)
```

Separating training and test data
---

```python
#splitting the dataset into test and train data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size =0.2, stratify = Y, random_state=3)
print('Summary Statistics of actual vs train vs test shape of the DataFrame are as follows:')
print('-'*100)
print('actual shape:',X.shape)
print('train shape:',X_train.shape)
print('test shape:',X_test.shape)
```

```python
#Logistic regression for model training with training data
model = LogisticRegression()
model.fit(X_train, Y_train)
```

Model Evaluation

```python
#accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
print('Accuracy on training data:', training_data_accuracy)
```

```python
#accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print('Accuracy on test data:', test_data_accuracy)
```
